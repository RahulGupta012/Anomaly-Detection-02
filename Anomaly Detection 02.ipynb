# Q10. How can global outliers be detected using the Isolation Forest algorithm?

Isolation Forest algorithem is based on the Decision tree technique, which is a best method for detecting the global outliers. To detect these outliers such steps are taking palce ;

**Isolation Tree Construction :**
By this algorithem a isolation tree has constructed in which each node is split in their two subnodes and the leaf node will cease the process from their end.

**Path length calculation :**
or each instance in the dataset, the algorithm measures the average path length required to isolate that instance across all trees. The path length represents the number of edges traversed from the root to the leaf node containing the instance.

**Anomaly Score Calculation:**
An anomaly score is calculated based on the average path length. Instances with shorter average path lengths are considered anomalies, as they required fewer splits to be isolated in the trees.

**Thresholding :**
Anomalies are identified by comparing the anomaly scores to a predefined threshold. Instances with scores below the threshold are considered outliers.

**Output:**
The algorithm outputs a binary label indicating whether each instance is an outlier (1) or not (0).


# Q11. What are some real-world applications where local outlier detection is more appropriate than global outlier detection, and vice versa?

**Local Outlier Detection:**

Applications: Network intrusion detection, manufacturing quality control, health monitoring, spatial anomaly detection.
Characteristics: Focuses on anomalies within specific regions or localized behavior.

**Global Outlier Detection:**

Applications: Financial fraud detection, credit scoring, quality assurance in manufacturing, population health monitoring.
Characteristics: Identifies anomalies across the entire dataset, suitable for detecting patterns that span a broader population or system.
